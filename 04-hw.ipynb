```python
import numpy as np


def build_hamming_matrix(m):
    n = (1 << m) - 1
    H = np.zeros((m, n), dtype=int)
    for col in range(n):
        val = col + 1
        for row in range(m):
            H[row, col] = (val >> row) & 1
    return H


def generate_simplex_words(H):
    m, n = H.shape
    words = []
    for mask in range(1 << m):
        word = np.zeros(n, dtype=int)
        for row in range(m):
            if (mask >> row) & 1:
                word = (word + H[row]) % 2
        words.append(word)
    return words


def check_simplex(m):
    H = build_hamming_matrix(m)
    words = generate_simplex_words(H)

    n = (1 << m) - 1
    expected_weight = 1 << (m - 1)

    print(f"Параметр m  {m}")
    print(f"Длина кода n  {n}")
    print(f"Ожидаемый вес  {expected_weight}")
    print()

    weights = []
    for w in words:
        wt = int(np.sum(w))
        weights.append(wt)

    nonzero_weights = [w for w in weights if w > 0]

    print("Веса ненулевых слов")
    print(nonzero_weights)
    print()

    all_equal = all(w == expected_weight for w in nonzero_weights)

    if all_equal:
        print("Свойство симплексного кода подтверждено")
    else:
        print("Свойство симплексного кода не подтверждено")

    print()
    print("Проверка завершена")
    print()


if __name__ == "__main__":
    for m in range(2, 5):
        check_simplex(m)
```
